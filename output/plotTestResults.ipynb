```python
data = np.loadtxt("testIntegrator1.txt", delimiter=",")

fig, ax = plt.subplots(nrows=2, ncols=2)

x = np.linspace(0, 0.7, 50)
ax[0, 0].set_title("Chute libre, v0!=0")
ax[0, 0].plot(x, 1. + (2.-9.81*0.01/2.)*x - (9.81/2.)*x*x)
ax[0, 0].scatter(data[:,0], data[:,1])
ax[0, 1].set_title("Chute verticale, v0=0")
ax[0, 1].scatter(data[:,3], data[:,4])
ax[1, 0].set_title("Chute verticale, v0=(0,2,0)")
ax[1, 0].scatter(data[:,6], data[:,7])
ax[1, 1].set_title("Chute libre, λ=0.3")
ax[1, 1].scatter(data[:,9], data[:,10])

plt.show()
```

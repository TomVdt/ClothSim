```python
dataReal = np.loadtxt("integrateur1ValeursReelles.txt", delimiter=" ")
dataEuler = np.loadtxt("testIntegrator1.txt", delimiter=",")
dataRK4 = np.loadtxt("testRK4Integrator.txt", delimiter=",")
dataNewmark = np.loadtxt("testNewmarkIntegrator.txt", delimiter=",")

fig, ax = plt.subplots()

ax.set_title("Chute libre, v0=(1, 2, 0")
ax.plot(dataReal[:,0], dataReal[:,2], label="Réelle", alpha=0.8)
ax.plot(dataEuler[:,0], dataEuler[:,1], c='r', label="Euler Cromer", alpha=0.5)
ax.plot(dataRK4[:,0], dataRK4[:,1], c='m', label="RK4", alpha=0.3)
ax.plot(dataNewmark[:,0], dataNewmark[:,1], c='b', label="Newmark", alpha=0.3)
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()

ax.set_title("Chute verticale, v0=0")
ax.scatter(dataEuler[:,3], dataEuler[:,4], c='r', s=2, label="Euler Cromer", alpha=0.5)
ax.scatter(dataRK4[:,3], dataRK4[:,4], c='m', s=2, label="RK4", alpha=0.3)
ax.scatter(dataNewmark[:,3], dataNewmark[:,4], c='b', s=2, label="Newmark", alpha=0.3)
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()

ax.set_title("Chute verticale, v0=(0,2,0)")
ax.scatter(dataEuler[:,6], dataEuler[:,7], c='r', s=2, label="Euler Cromer", alpha=0.5)
ax.scatter(dataRK4[:,6], dataRK4[:,7], c='m', s=2, label="RK4", alpha=0.3)
ax.scatter(dataNewmark[:,6], dataNewmark[:,7], c='b', s=2, label="Newmark", alpha=0.3)
ax.legend()

plt.show()
```
